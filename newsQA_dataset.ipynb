# **Split the dataset into train, validation, and test sets**

```python
from datasets import Dataset, DatasetDict

# Assuming 'newsqa_df' is your pandas DataFrame containing the NewsQA dataset

# Split the dataset into train, validation, and test sets
train_ratio = 0.8
val_ratio = 0.1
test_ratio = 0.1

train_size = int(train_ratio * len(newsqa_df))
val_size = int(val_ratio * len(newsqa_df))
test_size = len(newsqa_df) - train_size - val_size

train_data = newsqa_df[:train_size]
val_data = newsqa_df[train_size:train_size + val_size]
test_data = newsqa_df[train_size + val_size:]

# Convert the data into Dataset objects
train_dataset = Dataset.from_pandas(train_data)
val_dataset = Dataset.from_pandas(val_data)
test_dataset = Dataset.from_pandas(test_data)

# Create a DatasetDict
raw_datasets = DatasetDict({
    "train": train_dataset,
    "validation": val_dataset,
    "test": test_dataset
})

# Display the DatasetDict
print(raw_datasets)
```

```python
print("Context: ", raw_datasets["train"][0]["story_id"])
print("Question: ", raw_datasets["train"][0]["question"])
print("Answer: ", raw_datasets["train"][0]["validated_answers"])
```
