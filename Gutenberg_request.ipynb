```python
import requests

def get_book(book_id):
    url = f"http://www.gutenberg.org/cache/epub/{book_id}/pg{book_id}.txt"
    response = requests.get(url)
    if response.status_code == 200:
        return response.text
    else:
        return f"Failed to retrieve book {book_id}"

# Example usage:
book_id = 1342  # This is the book ID for "Pride and Prejudice" by Jane Austen
book_text = get_book(book_id)
print(book_text[:5000])  # Print the first 500 characters of the book
```
